```python
import csv

fifa_file = open('Fifa20.csv', encoding='utf-8')
file_reader = csv.reader(fifa_file)
player_data = list(file_reader)
fifa_file.close()

header = player_data[0]
player_data = player_data[1:]
for row in player_data:
    for idx in [2,3,4,7,8,12]:
        row[idx] = int(row[idx])
```

```python
print(header)
for row in player_data[:5]:
    print(row)
```

```python
#q1 What is the name of the oldest player?
def player(col,col2):
    val_play = 0
    for i in range(len(player_data)):
        if val_play < player_data[i][header.index(col)]:
            val_play = player_data[i][header.index(col)]
            j = i
    return player_data[j][header.index(col2)]
player("Age","Name")
```

```python
#q2 What is the name of the highest-paid player?
player("Wage","Name")
```

```python
#q3 What is the name of the highest valued player?
player("Value","Name")
```

```python
#q4 What is the nationality of the highest valued player?
player("Value","Nationality")
```

```python
#q5 What are the first five nationalities listed in the dataset?
def get_column(col_name):
    names = []
    for i in range(len(player_data)):
        names.append(player_data[i][header.index(col_name)])
    return names
get_column("Nationality")[:5]
```

```python
#q6 Which five player names are alphabetically first in the dataset?
alph = get_column("Name")
alph.sort()
alph[:5]
```

```python
#q7 What is the average Value?
def avg_val(t):
    total = 0
    for i in range(len(player_data)):
        if player_data[i][header.index(t)] > 0:
            total += player_data[i][header.index(t)]
                 
    return round(total/len(player_data),2)
avg_val("Value")        
```

```python
#q8 What is the average Height of the players?
avg_val("Height(cm)")        
```

```python
def count_players(p1,p2):
    p_pos = 0
    for i in range(len(player_data)):
        if player_data[i][header.index(p1)] == p2:
            p_pos+=1
    return p_pos
```

```python
#q9 How many players play in the position 'CAM'?
count_players("Player_Position","CAM")
```

```python
#q10 How many players are there per nationality?
def diff_type(typ): 
    per = {} 
    for i in get_column(typ): 
        per[i] = per.get(i, 0) + 1
    return per 

diff_type("Nationality")
```

```python
#q11 Which nationality has the most players participating in FIFA20?
xval = diff_type("Nationality")

tval = 0
country = ""
for key,val in xval.items():
    if val > tval:
        tval = val 
        country = key
country
```

```python
#q12 How many players have Argentina as their nationality?
count_players("Nationality","Argentina")
```

```python
#creating a new function player_to_dict
def player_to_dict(player_id):
    player_dict = {}
    for i in range(len(player_data)):
        if player_id == player_data[i][header.index("ID")]:
            player_dict = {
                'ID': player_data[i][header.index("ID")],
                'Name': player_data[i][header.index("Name")],
                 'Age': player_data[i][header.index("Age")],
                'Height(cm)': player_data[i][header.index("Height(cm)")],
                'Weight(kg)': player_data[i][header.index("Weight(kg)")],
                 'Nationality': player_data[i][header.index("Nationality")],
                 'Club': player_data[i][header.index("Club")],
                 'Value': player_data[i][header.index("Value")],
                 'Wage': player_data[i][header.index("Wage")],
                 'Player_Position': player_data[i][header.index("Player_Position")],
                 'Preferred_Foot': player_data[i][header.index("Preferred_Foot")],
                 'Body_Type': player_data[i][header.index("Body_Type")],
                 'Stamina': player_data[i][header.index("Stamina")]
            }
    return player_dict
```

```python
#q19 What is the average stamina of the club 'FC Bayern München' (see column Stamina)?

stm = 0
ct = 0
for i in range(len(player_data)): 
    if player_data[i][header.index("Club")] == "FC Bayern München":
        stm += player_data[i][header.index("Stamina")] 
        ct += 1
round(stm/ct,2)
```

```python
#function created to use for q20
def wage_club(player_data):
    w = 0 
    for i in range(len(player_data)):
        w += player_data[i][header.index("Wage")]
    return round(w/len(player_data),2)
wage_club(player_data)
```

```python
#q20 What is the average wage per club?

#the code has been used with refrence from mike's lecture
buckets = {}
for line in player_data:
    club = line[header.index("Club")]
    if not club in buckets:
        buckets[club] = []
    buckets[club].append(line)
    
averages = {}
for club in buckets:
    buckets[club]
    avg = wage_club(buckets[club])
    averages[club] =  avg
averages
```
